```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

W = 100
H = 100
im = np.zeros((W, H, 3), np.uint8) # La tercera dimensi√≥n (3) representa los canales de color (rojo, verde, azul).
# np.uint8 day un rengo 0 a 225 en un tama√±o 8 bits con poco resolucion

# Dibuja un cuadrado azul en el centro del cuadrado verde
for i in range(20, 80):
    for j in range(20, 80):
        im[i, j] = (0, 0, 255) # azul

# Dibuja un cuadrado verde en el centro del cuadrado rojo
for i in range(30, 70):
    for j in range(30, 70):
        im[i, j] = (0, 255, 0) # verde

# Dibuja un cuadrado rojo en el centro de la imagen
for i in range(40, 60):
    for j in range(40, 60):
        im[i, j] = (255, 0, 0) # Rojo

# Dibuja un cuadrado negr0 en el centro del cuadrado verde
for i in range(45, 55):
    for j in range(45, 55): # dimension del cuadro
        im[i, j] = (0, 0, 0) # Negro

cv2_imshow(im)
```

image.png

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

W = 600
H = 200

im = np.zeros((H, W), np.uint8)

# Definir el n√∫mero de l√≠neas horizontales (y su frecuencia) basadas en la funci√≥n seno
num_lines = 5
frequency = 40  # Frecuencia de la onda seno

# Dibujar las l√≠neas horizontales que siguen una funci√≥n seno
for i in range(num_lines):
    amplitude = int(H / (30))  # La amplitud de la l√≠nea var√≠a con la frecuencia
    y_offset = H // (num_lines + 1) * (i + 1)

    # Generar puntos basados en la funci√≥n seno
    x = np.arange(0, W)
    y = amplitude * np.sin(2 * np.pi * frequency * x / W) + y_offset

    # Convertir los puntos a enteros
    points = np.column_stack((x, y)).astype(np.int32)

    # Dibujar la l√≠nea
    cv2.polylines(im, [points], isClosed=False, color=255, thickness=1)

# Mostrar la imagen
cv2_imshow(im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
